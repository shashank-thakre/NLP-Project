```python
import pandas as pd
import spacy
import nltk
from collections import Counter
#spacy.require_gpu()
```

```python
df_news = pd.read_csv('df5.csv')
df_news.head()
```

```python
df_news.shape
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
%%time
all_orgs = []
all_person = []
for txt in df_news.combined:
    doc = nlp(txt)
    
    for ent in doc.ents:
        if ent.label_ == 'ORG':
            all_orgs.append(ent.text)
        elif ent.label_ == 'PERSON':
            all_person.append(ent.text)
```

```python
count1 = Counter(all_orgs)
count2 = Counter(all_person)
```

```python
df1 = pd.DataFrame(count1.most_common(20), columns = ['Org Name', 'Count'])
```

```python
df2 = pd.DataFrame(count2.most_common(20), columns = ['Person Name', 'Count'])
```
