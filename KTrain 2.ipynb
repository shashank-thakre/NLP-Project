```python
def regex_filter(val):
    exp2 = '[a-zA-Z]+ [a-zA-Z]+, \d+, \S+ Illinois'
    exp3 = '[a-zA-Z]+ [a-zA-Z]+, \d+, [a-zA-Z]+ [a-zA-Z]+, Illinois'
    mo = re.search(exp2,val)
    mo1 = re.search(exp2,val)
    if mo:
        return False
    elif mo1:
        return False
    else:
        return True
```

```python
import ktrain
import tensorflow as tf
import sys

print('Python Version: ' + sys.version)
print('TensorFlow Version: ' + tf.__version__)
```

```python
# Get texts in a list for ktrain
texts = df2['combined'].tolist()
```

```python
df2.head()
```

```python
%%time
tm = ktrain.text.get_topic_model(
    texts=texts, 
    n_topics=50, 
    n_features=10000, 
    min_df=5, 
    max_df=0.5, 
    stop_words='english', 
    model_type='lda', 
    lda_max_iter=5, 
    verbose=1)
```

```python
tm.print_topics()
```

```python
%%time
#tm.build(texts, threshold=0.25)
tm.build(texts)
```

```python
def get_topics_articles(idx):
#     print(idx)
#     print([idx])
    try:
        return np.argmax(tm.get_doctopics(doc_ids=[idx]))
    except:
        return 99
```

```python
%%time
df2['topics1'] = df2.apply(lambda x: get_topics_articles(x['index']), axis=1)
```
