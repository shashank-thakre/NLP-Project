```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

pd.set_option('display.max_colwidth', None)
```

### Read the previously saved file with sentiment scores

```python
df_news = pd.read_csv('Sentiment_score1.csv')
df_news.head(1)
```

### NER Orgs and Persons that want to check Sentiment for are:

#### Orgs:   
Chicago Police Department/CPD, Amazon  
#### Persons:  
Lori Lightfoot, Donald Trump, Pritzker  

```python
## Get data for Chicago Police Department/CPD

cpd_data = df_news[(df_news['combined'].str.contains('Chicago Police Department') | \
         df_news['combined'].str.contains('Chicago police department')| \
         df_news['combined'].str.contains('chicago police department')| \
         df_news['combined'].str.contains('Chicago Police')| \
         df_news['combined'].str.contains('Chicago police') | \
         df_news['combined'].str.contains('chicago police') |
         df_news['combined'].str.contains('CPD') | \
         df_news['combined'].str.contains('cpd')
        )]
```

```python
plt.rcParams["figure.figsize"] = (25,10)

tweet_timeline_df = cpd_data.copy()
tweet_timeline_df = tweet_timeline_df[['date', 'sentiment_score']].set_index('date')

tweet_timeline_df.groupby(by='date').mean().plot()
#tweet_timeline_df.plot()
plot = plt.title(f'Sentiment timeline')
```

```python
plt.rcParams["figure.figsize"] = (10,6)
cpd_data['sentiment_score'].value_counts().plot(kind='bar')
```

As per the sentiment plot above, most of the data about the police department has negative sentiments
