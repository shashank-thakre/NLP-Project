### K-Train

```python
# Get texts in a list for ktrain
texts = df_news['combined'].tolist()
```

```python
len(texts)
```

```python
%%time
tm = ktrain.text.get_topic_model(
    texts=texts, 
    n_topics=50, 
    n_features=10000, 
    min_df=5, 
    max_df=0.5, 
    stop_words='english', 
    model_type='lda', 
    lda_max_iter=5, 
    verbose=1)
```

```python
%%time
#tm.build(texts, threshold=0.25)
tm.build(texts)
```

```python
tm.print_topics(show_counts=True)
```

```python
def get_topics_articles(idx):
    return np.argmax(tm.get_doctopics(doc_ids=[idx]))
```

```python
%%time
df_news['topics1'] = df_news.apply(lambda x: get_topics_articles(x['index']), axis=1)
```
